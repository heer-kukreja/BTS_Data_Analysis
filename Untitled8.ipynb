# Cleaning data

```python
import re

def clean_data(text):
    letters_only = re.sub("[^a-zA-Z7]", " ", text) 
    wor = letters_only.lower().split()                            
    return( " ".join( wor ))

import nltk
nltk.download('words')
words = set(nltk.corpus.words.words())

bts = ['namjoon', 'taehyung', 'jungkook', 'jhope', 'seokjin', 'borahe', 'jin', 'seok', 'kim', 'jeon', 'jimin', 'tae', 'hyung', 'bangtan', 'bts', 'yoongi', 'jung', 'hoseok', '7', 'hobi', 'suga', 'v'] 
def eng_only(text):
  return(" ".join(w for w in nltk.wordpunct_tokenize(text) if w.lower() in words or not w.isalpha() or w in bts))
```

```python
comment_clean = df_youtube['comment_text'].apply(clean_data)
comment_clean = comment_clean.apply(eng_only)
```

```python
from wordcloud import WordCloud, STOPWORDS
stopwords= set(STOPWORDS)|{'a', 'and','the','more', 'it','they', 'them', 'are' 'has', 'have','will', 'be'}

def create_cloud(comment_clean):
    x= [i for i in comment_clean]
    y= ' '.join(x)
    cloud = WordCloud(background_color='white',width=1600, height=800,max_words=500,stopwords=stopwords).generate(y)
    plt.figure(figsize=(30,15))
    plt.axis('off')
    plt.imshow(cloud)
    plt.show()
```

```python
create_cloud(comment_clean)
```

# Verifying the Word Cloud with the bar graph


```python
comm = ' '.join(comment_clean.str.lower())
gb = ['m', 'n', 't', 's', 'even', 'always', 'y']
filtered_words = [word for word in comm.split() if word not in stopwords and word not in gb]
counted_words = collections.Counter(filtered_words)

words = []
counts = []
for letter, count in counted_words.most_common(10):
    words.append(letter)
    counts.append(count)
```

```python
colors = cm.rainbow(np.linspace(0, 1, 10))
rcParams['figure.figsize'] = 20, 10

plt.title('Top words in the comments vs their count')
plt.xlabel('Count')
plt.ylabel('Words')
plt.barh(words, counts, color=colors)
```

# Color Coded POS tagging based Word Cloud

```python
import collections
from nltk import pos_tag_sents
from nltk.corpus import reuters

import nltk
nltk.download('reuters')
```

```python
!unzip /root/nltk_data/corpora/reuters.zip -d /root/nltk_data/corpora
```

```python
tagged_sents = pos_tag_sents(reuters.sents())
 
word_tag_counters = collections.defaultdict(collections.Counter)
for tagged_sent in tagged_sents:
    for w, t in tagged_sent:
        word_tag_counters[w.lower()][t] += 1
```

```python
def get_color(word, **kwargs):
    # Get most common tag for the given word
    try:
        most_common_tag = word_tag_counters[word].most_common(1)[0][0]
    except:
        return '#111c63'

    # Red for Nouns
    if most_common_tag.startswith('NN'):
        return '#62108f'
 
    # Blue for Verbs
    elif most_common_tag.startswith('VB'):
        return '#4ae0e8'
 
    # Orange for Adjectives
    elif most_common_tag.startswith('JJ'):
        return 'orange'
 
    # Green for Adverbs
    elif most_common_tag.startswith('RB'):
        return '#7cfca2'

    # Gray for everything else
    return 'gray'
```

```python
wc = WordCloud(background_color='white', colormap='jet', max_words=100, width=800, height=400).generate(' '.join(comment_clean))
wc.recolor(color_func=get_color)
```

```python
plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
plt.show()
```

# Word Cloud of Adjectives

```python
from textblob import TextBlob

def get_adjectives(text):
    blob = TextBlob(text)
    return [ word for (word,tag) in blob.tags if tag == "JJ"]

comm = get_adjectives(comm)
```

```python
adj = " ".join(comm)
```

```python
wordcloud = WordCloud(width = 1000, height = 800, random_state=1, background_color='navy', colormap='rainbow', collocations=False, stopwords = STOPWORDS).generate(adj)
plt.figure(figsize=(40, 30))
# Display image
plt.imshow(wordcloud) 
# No axis details
plt.axis("off");
```
